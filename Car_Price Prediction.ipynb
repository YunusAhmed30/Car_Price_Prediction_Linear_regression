```python
data_selected = data[['price', 'fueltype', 'aspiration','carbody', 'drivewheel','wheelbase',
                  'curbweight', 'enginetype', 'cylindernumber', 'enginesize', 'boreratio','horsepower', 
                    'fueleconomy', 'carlength','carwidth', 'CarRange']]
data_selected.head()
```

```python
sns.pairplot(data_selected)
```

```python
#Creating dummy variables

def dummies(x,df):
    temp = pd.get_dummies(df[x], drop_first = True)
    df = pd.concat([df, temp], axis = 1)
    df.drop([x], axis = 1, inplace = True)
    return df


data_selected= dummies('fueltype',data_selected)
data_selected= dummies('aspiration',data_selected)
data_selected= dummies('carbody',data_selected)
data_selected= dummies('drivewheel',data_selected)
data_selected= dummies('enginetype',data_selected)
data_selected= dummies('cylindernumber',data_selected)
data_selected= dummies('CarRange',data_selected)
```

```python
#Train-test split and scaling

from sklearn.model_selection import train_test_split
df_train,df_test=train_test_split(data_selected,test_size=0.3,random_state=1)
```

```python
#Scaling features

from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
a=['price','wheelbase','curbweight','enginesize','boreratio','horsepower','fueleconomy','carlength','carwidth']
df_train[a]=scaler.fit_transform(df_train[a])
```

```python
y_train=df_train.pop('price')
x_train=df_train
```

```python
#Model building

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

```

```python
lm=LinearRegression()
lm.fit(x_train,y_train)
rfe=RFE(lm,10)
```

```python
rfe
```

```python
rfe=rfe.fit(x_train,y_train)
```

```python
list(zip(x_train.columns,rfe.support_,rfe.ranking_))
```

```python
x_train.columns[rfe.support_] #It will return the columns which has support as "true"
```

```python
x_train_rfe=x_train[x_train.columns[rfe.support_]]
x_train_rfe.head()
```

```python
def model_build(X,y):
    X=sm.add_constant(X)
    lm=sm.OLS(y,X).fit()
    print(lm.summary())
    return X

def checkVIF(X):
    vif=pd.DataFrame()
    vif['Features']=X.columns
    vif['VIF']=[variance_inflation_factor(X.values,i) for i in range(X.shape[1])]
    vif['VIF']=round(vif['VIF'],2)
    vif=vif.sort_values(by='VIF',ascending=False)
    return(vif)
```

```python
#model 1

x_train_new=model_build(x_train_rfe,y_train)
```

```python
checkVIF(x_train_new)
```

```python
#dropping curbweight because of high VIF value. (shows that curbweight has high multicollinearity.)
```

```python
x_train_new=x_train_new.drop(['curbweight'],axis=1)
```

```python
x_train_new.head()
```

```python
#Model 2
x_train_new=model_build(x_train_new,y_train)
```

```python
#p-vale of boreratio seems to be higher than the significance value of 0.05, 
#hence dropping it as it is insignificant in presence of other variables.
```

```python
x_train_new=x_train_new.drop(['boreratio'],axis=1)
```

```python
x_train_new.head()
```

```python
#Model 3
x_train_new=model_build(x_train_new,y_train)
```

```python
#p-vale of twelve seems to be higher than the significance value of 0.05,
#hence dropping it as it is insignificant in presence of other variables.
```

```python
x_train_new=x_train_new.drop(['twelve'],axis=1)
```

```python
#Model 4
x_train_new=model_build(x_train_new,y_train)
```

```python
checkVIF(x_train_new)
```

```python
# Residual analysis

lm=sm.OLS(y_train,x_train_new).fit()
y_train_price=lm.predict(x_train_new)
```

```python
y_train_price.head()
```

```python
#plotting the histogram of error terms

fig=plt.figure()
sns.distplot((y_train-y_train_price),bins=20)
fig.suptitle('Error terms',fontsize=30)
plt.xlabel('Errors')
```

```python
#Error terms seem to be approximately normally distributed, so the assumption on the linear modeling seems to be fulfilled.
```
